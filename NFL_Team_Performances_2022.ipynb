```python
#Calculate the Team Passing EPA Performances

#Mean EPA
mean_epa = pbp_summary_table.groupby('posteam').mean('epa').reset_index('posteam')[['posteam', 'epa']]
mean_epa = mean_epa.rename(columns={'posteam':'recent_team', 'epa':'mean_overall_epa_per_play'})
mean_epa['season'] = 2022
mean_epa.sort_values('mean_overall_epa_per_play', ascending = False)

#Median EPA
median_epa = pbp_summary_table.groupby('posteam').median('epa').reset_index('posteam')[['posteam', 'epa']]
median_epa = median_epa.rename(columns={'posteam':'recent_team', 'epa':'median_overall_epa_per_play'})
median_epa['season'] = 2022
median_epa.sort_values('median_overall_epa_per_play', ascending = False)

#Join the Tables
overall_epa = mean_epa.set_index(['recent_team', 'season']).join(median_epa.set_index(['recent_team', 'season']), on = ['recent_team', 'season'], how = 'outer', lsuffix = '_mean', rsuffix = '_median').reset_index(['recent_team', 'season'])
overall_epa.sort_values('mean_overall_epa_per_play', ascending = False)
```
